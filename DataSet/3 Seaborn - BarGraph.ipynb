## barplot() - Barplot / Bar Graph 

- barplot() shows the relation between a categorical variable and a continuous variable. 
- The data is represented in rectangular bars where the length the bar represents the proportion of the data in that category.
- Bar plot represents the estimate of central tendency. 

```python
import seaborn as sns 
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
sns.barplot(x = "sex", y = "survived", data = df)
plt.show()
df.head(3)
```

```python
# Pass dataset using data parameter
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex',data = tips_df)
```

```python
# modify the order of day
order = ['Sun', 'Thur', 'Fri', 'Sat']

sns.barplot(x = 'day', y = 'total_bill', hue = 'sex',
           data = tips_df, order = order)
```

```python
#Modify hue order
hue_order = ['Female', 'Male']


sns.barplot(x = 'day', y = 'total_bill', hue = 'sex',
           data = tips_df, hue_order = hue_order)
```

```python
# estimate y variable value and then plot
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex',
           data = tips_df, estimator= np.mean)
```

```python
# estimate y variable value and then plot
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex',
           data = tips_df, estimator= np.max)
```

```python
# set error bar size
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex',
           data = tips_df, ci = 20)
```

```python
# set error bar 
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex',
           data = tips_df,  n_boot=5)
```

```python
# Set barplot as vertical
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex',
           data = tips_df, orient='v')
```

```python
# set barplot as horizontal
sns.barplot(y = 'day', x = 'total_bill', hue = 'sex',
           data = tips_df,)
```

```python
# set barplot as horizontal
sns.barplot(x ='total_bill', y = 'size', hue = 'sex', data = tips_df,
           orient='h')
```

```python
# set color
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex',
           data = tips_df, color="g")
```

```python
# set color map or palette
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex',
           data = tips_df, palette="magma")
```

```python
# set saturation of barplot
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex',
           data = tips_df, saturation=.9)
```

```python
#set color of error bar
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex',
           data = tips_df, errcolor='0.9')
```

```python
# set width of error bar
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex',
           data = tips_df, errwidth= 12)
```

```python
# set cap size of error bar
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex',
           data = tips_df, capsize=0.1)
```

```python
# shift hue categorical variable bar bar in one bar
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex',
           data = tips_df, dodge= True)
```

```python
sns.set()
kwargs = {'alpha':0.9, 'linestyle':'-', 'linewidth':3, 'edgecolor':'b'}
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex',
           data = tips_df, **kwargs)
```

```python
# Pass Keyword  argument as parameter
sns.barplot(x = 'day', y = 'total_bill', 
           data = tips_df, alpha =.9, linestyle = "-.", linewidth = 3,
           edgecolor = "g")
```

```python
# Axes parameter
ax = sns.barplot(x = 'day', y = 'total_bill', 
           data = tips_df, alpha =.9, linestyle = "-.", linewidth = 3,
           edgecolor = "g")

ax.set(title = "Barplot of Tips DataFrame",
      xlabel = "Days",
      ylabel = "Total Bill")
```

```python
plt.figure(figsize = (16,9))

sns.barplot(x = 'day', y = 'total_bill', 
           data = tips_df, alpha =1, linestyle = "-.", linewidth = 3,
           edgecolor = "k")

plt.title("Barplot of Days and Total Bill", fontsize = 20)
plt.xlabel("Days", fontsize = 15)
plt.ylabel("Total Bill", fontsize = 15)

plt.savefig("Barplot of Days and Total Bill")
plt.show()
```

```python
data = tips_df
data.head(2)

data['sex'].value_counts().index
data['sex'].value_counts().values
```

```python
sns.set(style='whitegrid')

ax=sns.barplot(
                x=data['sex'].value_counts().index,
                y=data['sex'].value_counts().values,
                palette="Blues_d",
                hue=['Female','Male'])

plt.legend(loc=3)
plt.xlabel('Gender',color= 'r')
plt.ylabel('Frequency')
plt.title('Show of Gender Bar Plot')
plt.show()
```

```python
sns.barplot(x=data['gender'].value_counts().index,y=data['gender'].value_counts().values)
plt.title('Genders other rate')
plt.ylabel('Rates')
plt.legend(loc=0)
plt.show()
```

```python
plt.figure(figsize=(7,7))
sns.barplot(x=data['race/ethnicity'].value_counts().index,
              y=data['race/ethnicity'].value_counts().values)
plt.xlabel('race/ethnicity')
plt.ylabel('frequency')
plt.title('Show of Race/Ethnicity Bar Plot')
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.barplot(x = "parental level of education", y = "writing score", hue = "gender", data = data)
plt.xticks(rotation=45)
plt.show()
```

```python
sns.barplot(x = "parental level of education", y = "reading score", hue = "gender", data = data)
plt.xticks(rotation=45)
plt.show()
```

```python
sns.barplot(x = "parental level of education", y = "math score", hue = "gender", data = data)
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.catplot(y="gender", x="math score",
                 hue="parental level of education",
                 data=data, kind="bar")
plt.title('for Parental Level Of Education Gender & Math_Score')
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.catplot(x="gender", y="math score",
                 hue="test preparation course",
                 data=data, kind="bar")
plt.title('for Test Preparation Course Gender & Math_Score')
plt.show()
```

```python
ax = sns.barplot("parental level of education", "writing score", data=data,
                  linewidth=2.5, facecolor=(1, 1, 1, 0),
                  errcolor=".2", edgecolor=".2")
plt.xticks(rotation=90)
plt.show()
```

```python
data['test preparation course'].unique()
data_lunch_score=data[data['test preparation course']=="completed"].groupby(data['lunch'])['writing score'].sum()
plt.title("Lunch - Free/reduced & standard")
sns.barplot(x=data_lunch_score.index,y=data_lunch_score.values)
plt.show()
```

```python
f,ax=plt.subplots(figsize=(9,10))
sns.barplot(x=data['gender'].value_counts().values,y=data['gender'].value_counts().index,alpha=0.5,color='red',label='gender')
sns.barplot(x=data['race/ethnicity'].value_counts().values,y=data['race/ethnicity'].value_counts().index,color='blue',alpha=0.7,label='Race/Ethnicity')
ax.legend(loc='upper right',frameon=True)
ax.set(xlabel='gender , race/ethnicity',ylabel='Groups',title="Gender vs Race/Ethnicity ")
plt.show()
```
